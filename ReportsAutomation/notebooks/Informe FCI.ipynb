```python
from fpdf import FPDF
from datetime import datetime, timedelta
import os

from datetime import date
```

```python
WIDTH = 210
HEIGHT = 297

DATE = date.today()

encabezado = r"C:\Users\Bloomberg\Desktop\Gabriel\encabezado_fci.jpg"
logo = r"C:\Users\Bloomberg\Desktop\Gabriel\logo.png"
tabla_mm = r"C:\Users\Bloomberg\Desktop\Gabriel\fci_mm.png"
tabla_dl = r"C:\Users\Bloomberg\Desktop\Gabriel\fci_dl.png"
tabla_tomorrow = r"C:\Users\Bloomberg\Desktop\Gabriel\fci_t+1.png"
tabla_cer = r"C:\Users\Bloomberg\Desktop\Gabriel\fci_cer.png"
tabla_pyme = r"C:\Users\Bloomberg\Desktop\Gabriel\fci_py.png"

footer = r"C:\Users\Bloomberg\Desktop\Gabriel\footer2.png"

blanco = [255,255,255]

font = 'opensans'
```

```python
def create_title_fci(day, pdf):
    pdf.add_font(font,'', r"C:\Users\Bloomberg\Downloads\Open_Sans\OpenSans-Regular.ttf", uni=True)
    pdf.set_font(font, '', 13)  
    pdf.set_text_color(23, 49, 76)
    pdf.ln(25)
    pdf.write(5, f'{day}')
    pdf.set_font(font, '', 13)
    pdf.set_text_color(23, 49, 76)
    pdf.ln(7)
    pdf.write(5, f'FCI - MONEY MARKET')
    pdf.set_font(font, '', 13)  
    pdf.set_text_color(23, 49, 76)
    pdf.ln(115)
    pdf.write(5, f"FCI - T+1")
    pdf.set_font(font, '', 13)  
    pdf.set_text_color(23, 49, 76)
    pdf.ln(60)
    pdf.write(5, f"FCI - CER")
    
def create_title_fci2(day, pdf):
    pdf.add_font(font,'', r"C:\Users\Bloomberg\Downloads\Open_Sans\OpenSans-Regular.ttf", uni=True)
    pdf.set_font(font, '', 13)  
    pdf.set_text_color(23, 49, 76)
    pdf.ln(25)
    pdf.write(5, f'{day}')
    pdf.set_font(font, '', 13)
    pdf.set_text_color(23, 49, 76)
    pdf.ln(7)
    pdf.write(5, f'FCI - DÃ“LAR LINKED')
    pdf.set_font(font, '', 13)  
    pdf.set_text_color(23, 49, 76)
    pdf.ln(55)
    pdf.write(5, f"FCI - PYME")
   
    
def create_analytics_report(day=DATE, filename="Informe_FCI.pdf"):
    pdf = FPDF() # A4 (210 by 297 mm)
    
    pdf.add_page()
    pdf.image(encabezado, 0, 0, WIDTH)
    pdf.image(tabla_mm, 2.5, 50, WIDTH-5)
    pdf.image(tabla_tomorrow, 2.5, 165, WIDTH-5)
    pdf.image(tabla_cer, 2.5, 225, WIDTH-5)
    pdf.image(footer, 0, 285, WIDTH)
    create_title_fci(day, pdf)
    
    pdf.add_page()
    pdf.image(encabezado, 0, 0, WIDTH)
    pdf.image(tabla_dl, 5, 50, WIDTH-10)
    pdf.image(tabla_pyme, 5, 105, WIDTH-10)
    pdf.image(footer, 0, 285, WIDTH)
    create_title_fci2(day, pdf)
    pdf.output(filename, 'F')


if __name__ == '__main__':
    yesterday = (datetime.today() - timedelta(days=1)).strftime("%m/%d/%y").replace("/0","/").lstrip("0")
    yesterday = DATE 
  
    create_analytics_report(yesterday)
```
